# Task
Analyze and visualize the Titanic dataset using pandas and seaborn.

## Load the dataset

### Subtask:
Load the Titanic dataset from a publicly available source.


**Reasoning**:
Import pandas and load the Titanic dataset from the given URL into a DataFrame, then display the first 5 rows.



```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
display(df.head())
```

## Data preprocessing

### Subtask:
Perform necessary preprocessing steps such as handling missing values, encoding categorical variables, and creating new features.


**Reasoning**:
Handle missing values in the 'Age' and 'Embarked' columns, and drop the 'Cabin' column due to a high number of missing values. Create a 'FamilySize' feature and drop the original 'SibSp' and 'Parch' columns. Encode the 'Sex' and 'Embarked' columns using one-hot encoding. Display the first few rows of the modified DataFrame.



```python
import numpy as np

# Handle missing values in 'Age' by imputing with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Handle missing values in 'Embarked' by imputing with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column due to a large number of missing values
df.drop('Cabin', axis=1, inplace=True)

# Create 'FamilySize' feature
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Drop 'SibSp' and 'Parch' columns
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)

# Encode 'Sex' and 'Embarked' columns
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

display(df.head())
```

## Data visualization

### Subtask:
Visualize the data to explore relationships between features and the target variable (survival).


**Reasoning**:
Visualize the data to explore relationships between features and the target variable (survival) as per the instructions.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot of survivors vs non-survivors
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.xticks([0, 1], ['Did not survive', 'Survived'])
plt.title('Survival Count')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.show()

# Countplot of survival rate based on Sex
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex_male', hue='Survived', data=df)
plt.xticks([0, 1], ['Female', 'Male'])
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()

# Histogram of Age distribution by survival
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', kde=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()


# Box plot of Pclass and Fare
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   More passengers did not survive than survived.
*   A significantly higher proportion of female passengers survived compared to male passengers.
*   The Age distribution by survival plot suggested that younger age groups had a notable peak of non-survivors, while survivors were distributed across a broader range of ages.
*   Passengers in lower classes (Pclass 2 and 3) generally paid lower fares compared to those in the first class (Pclass 1).
